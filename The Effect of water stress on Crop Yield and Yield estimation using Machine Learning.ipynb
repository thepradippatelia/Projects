#### Splitting the dataframe into two dataframes, one includes input columns and  the second one includes target variable

```python
X = df.drop('Yield (Tonnes/Hectare)',axis=1)
y = df['Yield (Tonnes/Hectare)']
```

#### Splitting the datasets into train and test datasets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=32)
```

#### Now, in order to train the model, we have to encode our categorical variable.

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
X_train['District'] = encoder.fit_transform(X_train['District'])
X_test['District'] = encoder.transform(X_test['District'])
```

#### Features in our input dataset have different range of values, so we need to standarize them to bring them into same scale

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## `Linear Regression`

#### We need to first import the model

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

#### Fitting the model with train dataset

```python
LR.fit(X_train,y_train)
```

#### Checking the accuracy on train data itself as model is trained

```python
LR.score(X_train,y_train)
```

#### Prediction on test dataset

```python
y_pred = LR.predict(X_test)
```

#### Model performance on test dataset

```python
from sklearn.metrics import r2_score, mean_squared_error
print("r2_score:",r2_score(y_test,y_pred))
print("mean_squared_error:",mean_squared_error(y_test,y_pred))
```

# `Polynomial Linear Regression`

#### Need to first import the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

#### Creating polynomial features

```python
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

#### Fitting a linear regression model to the polynomial features

```python
reg = LinearRegression()
reg.fit(X_train_poly, y_train)
```

#### Prediction on test dataset

```python
y_pred = reg.predict(X_test_poly)
```

#### Model Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error
print("r2_score:",r2_score(y_test,y_pred))
print("mean_squared_error:",mean_squared_error(y_test,y_pred))
```

# `Decision Tree`

#### Need to first import the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DT = DecisionTreeRegressor()
```

#### we need tune the hyperparameters, so we must use GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'max_depth': [3, 5, 7],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 2, 4],
              'max_features': [None, 'sqrt', 'log2']}
```

```python
grid = GridSearchCV(DT,param_grid=param_grid,cv=5)
```

#### Fitting the model with train dataset

```python
grid.fit(X_train,y_train)
```

#### Model's  performance on train dataset

```python
grid.score(X_train,y_train)
```

#### Prediction on test dataset

```python
y_pred=grid.predict(X_test)
```

#### Model Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error
print("r2_score:",r2_score(y_test,y_pred))
print("mean_squared_error:",mean_squared_error(y_test,y_pred))
```
